```python
import numpy as np
import matplotlib.pyplot as plt

# Simulierte Daten generieren
np.random.seed(42)  # Für Reproduzierbarkeit
data = np.random.normal(loc=0, scale=1, size=1000)  # Normalverteilte Daten

# Histogramm erstellen
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogramm der simulierten Daten')
plt.xlabel('Wert')
plt.ylabel('Häufigkeit')
plt.grid(True)

# Histogramm anzeigen
plt.show()
```
